## **1.  Histogram Berapa Banyak Customer Berdasarkan Usia Yang Menggunakan Layanan Bank ini**

## **3.  Barplot Menghitung Jumlah Transaksi Yang Dilakukan Cutomer Berdasarkan Status Pernikahan**

```
totalMarital = []
maritalCount = df['Marital_Status'].unique() # mengkategorikan data
for x in maritalCount:
    dataCategory = df.loc[(df['Marital_Status']==x)] # menentukan data yang mana yang digunakan
    totalDataCategory = len(dataCategory) # Semua data dari kolom yang digunakan
    totalMarital.append(totalDataCategory)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 10))
plt.title("Rata-Rata Total Transaction Count Customer Berdasarkan Marital Status")
maritalCountBar = sns.barplot(x=maritalCount, y=totalMarital, palette = 'tab10')
# Untuk Menambahkan Angka diatas Setiap Bar
for i, d in enumerate(maritalCountBar.patches):
    x = d.get_x() + d.get_width() /2
    y = d.get_y() + d.get_height() + 0.5
    maritalCountBar.text(x, y, totalMarital[i],ha="center")
plt.show()
```

Tujuan dari diagram batang diatas adalah untuk mengetahui jumlah customer terhadap status pernikahannya

Pertama kita menentukan ada berapa status pernikahan yang ada di dalam data dan masing-masing status kita jumlahkan total transaksinya 

Dari diagram batang ini didapat kesimpulan bahwa Jumlah status pernikahan yang paling banyak yaitu pada status married yang memiliki jumlah 1189 dan jumlah terendah terdapat pada status Unknown yang memiliki jumlah 96

##**4. Barplot Pengguna Kredit Berdasarkan Education Level Dan Gendernya**

```
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,5))
plt.title('Barplot Pengguna Kredit Berdasarkan Education Level Dan Gender',fontsize=14)

ax = sns.countplot(x='Education_Level',hue='Gender', data=df, palette = 'mako')
ax.set(xlabel='Education Level', ylabel='Jumlah')
plt.show()
print(df['Education_Level'].value_counts())
```

Dari diagram batang ini dapat ditarik kesimpulan bahwa customer graduate paling banyak dengan jumlah 630 dan paling sedikit Doctorate yaitu 73. Pada setiap Education level, Gender Male lebih mendominasi daripada female.

##**5. Scatterplot Antara Total Transaction Amount dan Total Transaction Count**

```
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 10))
plt.title('Scatterplot Total Transaksi Dengan Total Pengeluaran Kredit', fontsize="14")

ax = sns.scatterplot(data=df, x="Total_Trans_Amt", y= "Total_Trans_Ct", color="green")
ax.set(xlabel="Total Transaction Amount", ylabel="Total Transaction Count")
plt.show()

correl = round(np.corrcoef(df["Total_Trans_Amt"],df["Total_Trans_Ct"])[1][0],2)
print("Korelasi antara keduanya : ",correl)

if(correl > 0.5 or correl < -0.5):
  if(correl > 0):
    print('Korelasi Total Transaction Amount dan Total Transaction Count  Adalah',correl,'\nMaka Korelasi Tersebut Merupkan Korelasi Yang Kuat Dan Merupakan Korelasi Positif')
  else:
    print('Korelasi Total Transaction Amount dan Total Transaction Count  Adalah',correl,'\nMaka Korelasi Tersebut Merupkan Korelasi Yang Kuat Dan Merupakan Korelasi Negatif')
else:
  if(correl > 0):
    print('Korelasi Total Transaction Amount dan Total Transaction Count  Adalah',correl,'\nMaka Korelasi Tersebut Merupkan Korelasi Yang Lemah Dan Merupakan Korelasi Positif')
  else:
       print('Korelasi Total Transaction Amount dan Total Transaction Count  Adalah',correl,'\nMaka Korelasi Tersebut Merupkan Korelasi Yang Lemah Dan Merupakan Korelasi Negatif')
```

correlation_coefficient.png

Berdasarkan ScatterPlot ini dapat ditarik kesimpulan bahwa Total Transaksi dengan Total Pengeluaran Kredit memiliki hubungan yang kuat. dapat dilihat dari scatterplot diatas karena titik mengelompok dalam bentuk linear yang jelas dan membentuk diagonal dari kiri kekanan makanya dapat kita prediksi bahwa korelasinya bernilai positif dan memiliki hubungan yang kuat. 

Untuk memastikannya kita lakukan pencarian korelasinya dan setelah ditemukan kita mendapaktan kesimpulan bahwa korelasi keduanya bernilai 0.81 , berdasarkan Correlation Coefficent diatas dapat ditarik kesimpulan bahwa korelasi antara keduanya memiliki hubungan yang kuat dan korelasinya bernilai Positif
